**3 Detecting outliers using Z-Score:**

Z-Score is also called a standard score. This value/score helps to understand how far the data point from
the mean is. And after setting up a threshold value one can utilize z score values of data points to define
the outliers.

Zscore = (data_point-mean) / std. deviation


```python
z = np.abs(stats.zscore(df['math_score']))
z
```

```python
threshold = 1.8
sample_outliers = np.where(z >threshold)
sample_outliers
```

```python
df['math_score']
```

**4 Detecting outliers using Inter Quantile
Range(IQR):**

IQR = Quartile3 – Quartile1 To define the outlier base value is defined above and below datasets normal
range namely Upper and Lower bounds, define the upper and the lower bound (1.5IQR value is
considered) : upper = Q3 +1.5IQR lower = Q1 – 1.5*IQR


```python
sorted_rscore= sorted(df['reading_score'])
sorted_rscore
```

```python
# Calculate and print Quartile 1 and Quartile 3
q1 = np.percentile(sorted_rscore, 25)
q3 = np.percentile(sorted_rscore, 75)
print(q1,q3)
```

```python
IQR = q3-q1
IQR
```

```python
lwr_bound = q1-(1.5*IQR)
upr_bound = q3+(1.5*IQR)
print(lwr_bound, upr_bound)
```

```python
#Print Outliers
r_outliers = []
for i in sorted_rscore:
 if (i<lwr_bound or i>upr_bound):
 r_outliers.append(i)
print(r_outliers)
```

```python
df['reading_score']
```

**Handling of Outliers:**

Below are some of the methods of treating the outliers 1.Trimming/removing the outlier 2.Quantile based
flooring and capping 3.Mean / Median imputation

**1.Trimming/removing the outlier**

```python
new_df=df.copy()
for i in sample_outliers:
 new_df.drop(i,inplace=True)
new_df
```

**2.Quantile based flooring and capping**

In this technique, the outlier is capped at a certain value above the 90th percentile value or floored at a
factor below the 10th percentile value. OR In this technique, we will do the flooring (e.g., the 10th
percentile) for the lower values and capping (e.g., the 90th percentile) for the higher values.


```python
df_stud=df.copy()
```
